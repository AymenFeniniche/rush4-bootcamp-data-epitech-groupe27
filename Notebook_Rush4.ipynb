## Feature Engineering

```python
df = data_camp_market.copy()

# 1) Âge (année courante pour rester robuste)
df["Age"] = 2014 - df["Year_Birth"]

# 2) Nombre total d’enfants
df["Children"] = df["Kidhome"] + df["Teenhome"]

# 3) Ancienneté client (en années, depuis aujourd’hui)
today = pd.Timestamp.today().normalize()
df["Customer_Since_Years"] = (today - df["Dt_Customer"]).dt.days / 365.25

# 4) Dépense totale
spend_cols = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds"]
df["Total_Spent"] = df[spend_cols].sum(axis=1)

# 5) Canal préféré (web / catalogue / magasin) selon le max des achats
channel_cols = ["NumWebPurchases","NumCatalogPurchases","NumStorePurchases"]
preferred_idx = df[channel_cols].idxmax(axis=1)
df["Preferred_Channel"] = preferred_idx.map({
    "NumWebPurchases": "web",
    "NumCatalogPurchases": "catalog",
    "NumStorePurchases": "store"
})

# 6) Nombre de campagnes acceptées
cmp_cols = ["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5"]
df["Campaign_Accepted"] = df[cmp_cols].sum(axis=1)

# 7) Taux de conversion web : achats web / visites web (évite /0)
df["Web_Conversion_Rate"] = df["NumWebPurchases"] / df["NumWebVisitsMonth"].replace(0, np.nan)
df["Web_Conversion_Rate"] = df["Web_Conversion_Rate"].fillna(0)

# 👉 Regarde vite fait le résultat
new_cols = ["Age","Children","Customer_Since_Years","Total_Spent","Preferred_Channel","Campaign_Accepted","Web_Conversion_Rate"]
df[new_cols].head()
```

## ACP - Analyse en Composantes Principales

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

# ==== Paramétrage (à remplacer par tes valeurs) ====
df = data_camp_market.copy()
num_vars = ["Income","Recency","MntWines","MntMeatProducts","MntFishProducts","MntFruits","MntSweetProducts","MntGoldProds"]
cat_vars = []            # ex: ["Education","Marital_Status"]
id_col  = "ID"           # ou None
n_components = None      # None => toutes, on choisira après via le scree plot

# ==== Sous-ensemble + gestion ID ====
cols = (num_vars or []) + (cat_vars or [])
work = df[cols].copy()

# Option simple : on droppe les lignes incomplètes pour démarrer
work = work.dropna()

# ==== Pipeline encodage + standardisation + PCA ====
ct = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(with_mean=True, with_std=True), num_vars),
        ("cat", OneHotEncoder(drop="if_binary", handle_unknown="ignore"), cat_vars)
    ],
    remainder="drop"
)

pca = PCA(n_components=n_components, random_state=42)

pipe = Pipeline(steps=[
    ("prep", ct),
    ("pca", pca)
])

X_pca = pipe.fit_transform(work)

# ==== Variance expliquée ====
explained = pipe.named_steps["pca"].explained_variance_ratio_
cumexp = np.cumsum(explained)

# Choix de k (exemple : premier k avec >= 85% de variance)
k = int(np.argmax(cumexp >= 0.85) + 1) if explained.size > 0 else 2

# Si tu veux refitter avec k retenu :
if k != pipe.named_steps["pca"].n_components_:
    pca = PCA(n_components=k, random_state=42)
    pipe = Pipeline(steps=[("prep", ct), ("pca", pca)])
    X_pca = pipe.fit_transform(work)
    explained = pipe.named_steps["pca"].explained_variance_ratio_
    cumexp = np.cumsum(explained)

# ==== Loadings (corrélations variable–composante) ====
# Récupérer les noms de features après encodage
feat_names_num = num_vars
feat_names_cat = []
if cat_vars:
    ohe = pipe.named_steps["prep"].named_transformers_["cat"]
    feat_names_cat = list(ohe.get_feature_names_out(cat_vars))
feat_names = feat_names_num + feat_names_cat

components = pipe.named_steps["pca"].components_          # shape (k, n_features)
explained_var = pipe.named_steps["pca"].explained_variance_
loadings = components.T * np.sqrt(explained_var)           # shape (n_features, k)

loadings_df = pd.DataFrame(loadings, index=feat_names,
                           columns=[f"PC{i+1}" for i in range(loadings.shape[1])])

# ==== Scores des individus ====
scores = pd.DataFrame(X_pca, index=work.index,
                      columns=[f"PC{i+1}" for i in range(X_pca.shape[1])])

# ==== (Option) joindre les scores au df d’origine ====
df_scores = scores.copy()
if id_col and id_col in df.columns:
    df_scores[id_col] = df.loc[work.index, id_col].values

# Tu peux maintenant tracer un scree plot et un biplot.
```

```python
import matplotlib.pyplot as plt

# Scree plot
plt.figure(figsize=(6,4))
plt.plot(range(1, len(explained)+1), explained, marker="o")
plt.plot(range(1, len(cumexp)+1), cumexp, marker="x")
plt.title("Scree plot: variance expliquée (barres) et cumulée (croix)")
plt.xlabel("Composante")
plt.ylabel("Proportion de variance")
plt.grid(True)
plt.show()

# Biplot rapide (PC1 vs PC2)
pcx, pcy = "PC1", "PC2"
plt.figure(figsize=(7,6))
plt.scatter(scores[pcx], scores[pcy], alpha=0.6)
# Vecteurs de variables (limiter aux top variables si beaucoup)
scale = 2.0
for var, (lx, ly) in loadings_df[[pcx, pcy]].iterrows():
    plt.arrow(0, 0, lx*scale, ly*scale, head_width=0.03, length_includes_head=True)
    plt.text(lx*scale*1.1, ly*scale*1.1, var, fontsize=8)
plt.axhline(0, color="grey", linewidth=0.8)
plt.axvline(0, color="grey", linewidth=0.8)
plt.xlabel(pcx); plt.ylabel(pcy)
plt.title("Biplot (PC1 vs PC2)")
plt.tight_layout()
plt.show()
```

```python
# # ===============================
# # K-means sur l’espace PCA
# # ===============================
# from sklearn.cluster import KMeans
# from sklearn.metrics import silhouette_score

# # ---------- 1) Choix de la base d’entrée (scores PCA) ----------
# # On prend les premières composantes expliquant ~80% de la variance si possible,
# # sinon on prend toutes les composantes disponibles.
# X_full = scores.copy()  # scores vient de l’étape PCA précédente (PC1, PC2, ...)
# if 'cumexp' in globals():
#     # Nombre minimal de composantes pour atteindre 80% de variance expliquée
#     target = 0.80
#     m = int(np.argmax(np.array(cumexp) >= target) + 1) if np.any(np.array(cumexp) >= target) else X_full.shape[1]
# else:
#     # Si cumexp n’est pas dispo, on garde par défaut les 5 premières (ou toutes si <5)
#     m = min(5, X_full.shape[1])

# X = X_full.iloc[:, :m].to_numpy()

# print(f"Nombre de composantes utilisées pour les k-means : {m}/{X_full.shape[1]}")

# # ---------- 2) Choix du nombre de clusters (Elbow + Silhouette) ----------
# k_min, k_max = 2, 10
# inertias = []
# silhouettes = []

# for k in range(k_min, k_max + 1):
#     km = KMeans(n_clusters=k, n_init=10, random_state=42)
#     labels_k = km.fit_predict(X)
#     inertias.append(km.inertia_)
#     # Silhouette non définie pour k=1; ici k>=2 donc OK, mais on sécurise si un problème survient
#     try:
#         sil = silhouette_score(X, labels_k)
#     except Exception:
#         sil = np.nan
#     silhouettes.append(sil)

# # Courbe "coude" (inertie)
# plt.figure(figsize=(6,4))
# plt.plot(range(k_min, k_max+1), inertias, marker="o")
# plt.xlabel("k")
# plt.ylabel("Inertie (within-cluster SSE)")
# plt.title("Elbow plot (k-means)")
# plt.grid(True)
# plt.show()

# # Courbe silhouette
# plt.figure(figsize=(6,4))
# plt.plot(range(k_min, k_max+1), silhouettes, marker="o")
# plt.xlabel("k")
# plt.ylabel("Score de silhouette")
# plt.title("Silhouette par k")
# plt.grid(True)
# plt.show()

# # Sélection automatique du k qui maximise la silhouette
# valid_sil = np.array(silhouettes, dtype=float)
# if np.all(np.isnan(valid_sil)):
#     # Si tout est NaN (très rare), fallback: k=4
#     best_k = 4
#     print("Avertissement: silhouette indisponible. On retient k=4 par défaut.")
# else:
#     best_k = int(np.nanargmax(valid_sil) + k_min)
# print(f"k retenu (max silhouette): {best_k}")

# # ---------- 3) Entraînement final et affectation des clusters ----------
# kmeans = KMeans(n_clusters=best_k, n_init=10, random_state=42)
# cluster_labels = kmeans.fit_predict(X)

# # Ajout au DataFrame des scores PCA
# scores_km = X_full.copy()
# scores_km["cluster"] = cluster_labels

# # Si tu veux aussi l’ajouter au df d’origine :
# df_km = df.copy() if 'df' in globals() else data_camp_market.copy()
# df_km = df_km.loc[scores_km.index]  # alignement des index individus
# df_km["cluster"] = cluster_labels

# # ---------- 4) Visualisation rapide sur PC1 vs PC2 ----------
# pcx, pcy = "PC1", "PC2"
# if pcx in X_full.columns and pcy in X_full.columns:
#     plt.figure(figsize=(7,6))
#     for c in sorted(scores_km["cluster"].unique()):
#         subset = scores_km[scores_km["cluster"] == c]
#         plt.scatter(subset[pcx], subset[pcy], label=f"cluster {c}", alpha=0.7)
#     plt.axhline(0, color="grey", linewidth=0.8)
#     plt.axvline(0, color="grey", linewidth=0.8)
#     plt.legend(title="Clusters")
#     plt.xlabel(pcx); plt.ylabel(pcy)
#     plt.title(f"K-means (k={best_k}) projeté sur {pcx} vs {pcy}")
#     plt.tight_layout()
#     plt.show()
# else:
#     print("PC1/PC2 non trouvées dans 'scores' — biplot clusters non affiché.")

# # ---------- 5) Profilage des clusters ----------
# # On calcule des moyennes par cluster sur quelques variables métiers pertinentes.
# vars_profil = []
# if 'num_vars' in globals():
#     vars_profil += list(num_vars)
# # Tu peux ajouter des variables d’intérêt spécifiques ici, par ex. :
# vars_profil = [v for v in vars_profil if v in df_km.columns]  # sécurisation

# if vars_profil:
#     profil = df_km.groupby("cluster")[vars_profil].mean().round(2)
#     display(profil)
# else:
#     print("Aucune variable numérique métier disponible pour le profilage (vérifie 'num_vars').")

# # ---------- 6) (Option) Sauvegardes pratiques ----------
# # scores_km.to_csv("scores_pca_kmeans.csv", index=False, sep=";")
# # df_km.to_csv("data_kmeans.csv", index=False, sep=";")

# # Objets de sortie utiles :
# # - best_k : nombre de clusters retenu
# # - kmeans : modèle KMeans entraîné
# # - df_km  : df original aligné + colonne 'cluster'
# # - scores_km : scores PCA + colonne 'cluster'


# === Segmentation clients par K-means (bloc prêt à coller) ===
# Prérequis: pandas, numpy, scikit-learn, matplotlib
# Si besoin: pip install pandas numpy scikit-learn matplotlib

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

RANDOM_STATE = 42

# 1) Récupération des données
if 'df' in globals() and isinstance(df, pd.DataFrame):
    _df_src = 'df (déjà présent dans le notebook)'
else:
    # essaie d'abord le fichier dans le dossier courant, sinon /mnt/data (utile sur certaines plateformes)
    candidate_paths = ['Camp_Market.csv', './Camp_Market.csv', '/mnt/data/Camp_Market.csv']
    csv_path = next((p for p in candidate_paths if os.path.exists(p)), None)
    if csv_path is None:
        raise FileNotFoundError("Impossible de trouver 'Camp_Market.csv'. Place-le à la racine du notebook.")
    df = pd.read_csv(csv_path, sep=';')
    _df_src = f'csv lu: {csv_path}'

print(f"[INFO] Source des données: {_df_src}")
print(f"[INFO] Dimensions: {df.shape}")

# 2) Feature engineering (variables utiles à la segmentation)
# - Dates en datetime + ténure (ancienneté)
df = df.copy()
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True, errors='coerce')

# Année de référence = plus récente date client connue (sinon 2014 par défaut, cohérent avec ce dataset)
ref_date = df['Dt_Customer'].max()
if pd.isna(ref_date):
    ref_date = pd.Timestamp('2014-12-31')

df['CustomerTenureDays'] = (ref_date - df['Dt_Customer']).dt.days

# - Démographie
df['Children'] = df['Kidhome'] + df['Teenhome']
df['Age'] = ref_date.year - df['Year_Birth']

# - Montants et comportements d’achat
amount_cols = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
purchase_cols = ['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']

df['TotalMnt'] = df[amount_cols].sum(axis=1)
df['TotalPurchases'] = df[purchase_cols].sum(axis=1)

# - Taux/ratio prudents
df['AvgSpendPerPurchase'] = df['TotalMnt'] / np.where(df['TotalPurchases']>0, df['TotalPurchases'], np.nan)

# - Engagement campagne (nombre de campagnes acceptées, y compris la réponse finale)
campaign_cols = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']
df['CampaignsAccepted'] = df[campaign_cols].sum(axis=1)

# - Statut de couple (binaire simple, utile pour profiling)
df['IsCouple'] = df['Marital_Status'].isin(['Married','Together']).astype(int)

# 3) Variables candidates pour le clustering (numériques, robustes et interprétables)
features = [
    'Age', 'Income', 'Children',
    'Recency',                  # plus petit = achat récent
    'TotalMnt', 'TotalPurchases',
    'NumWebVisitsMonth',
    'AvgSpendPerPurchase',
    'CampaignsAccepted',
    'CustomerTenureDays'
]

# Vérif de présence + conversion numérique prudente
present_feats = [c for c in features if c in df.columns]
if len(present_feats) < 4:
    raise ValueError(f"Assez peu de variables trouvées pour clusteriser. Trouvées: {present_feats}")

X = df[present_feats].apply(pd.to_numeric, errors='coerce')

# 4) Pipeline: imputation médiane + standardisation
pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

X_proc = pipe.fit_transform(X)

# 5) Choix de k: courbe elbow + silhouette (k de 2 à 10)
ks = range(2, 11)
inertias = []
silhs = []
models = {}

for k in ks:
    km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=20)
    labels = km.fit_predict(X_proc)
    models[k] = km
    inertias.append(km.inertia_)
    silhs.append(silhouette_score(X_proc, labels))

# Sélection via meilleur score silhouette (pratique pour un choix auto)
k_best = ks[int(np.argmax(silhs))]


print(f"[INFO] k choisi (silhouette max): {k_best:.0f} | silhouette={max(silhs):.3f}")

# Traces (Elbow & Silhouette)
plt.figure()
plt.plot(list(ks), inertias, marker='o')
plt.title("Méthode du coude (Inertia vs k)")
plt.xlabel("k"); plt.ylabel("Inertia"); plt.grid(True)
plt.show()

plt.figure()
plt.plot(list(ks), silhs, marker='o')
plt.title("Silhouette moyenne vs k")
plt.xlabel("k"); plt.ylabel("Silhouette"); plt.grid(True)
plt.show()

# 6) Entraînement final K-means avec k_best
kmeans = models[k_best]
labels = kmeans.labels_

# 7) Ajout des clusters au DF d’origine
df['cluster'] = labels

# 8) Profiling des segments: moyennes (et taille)
profile_cols = present_feats + ['IsCouple']
cluster_profile = df.groupby('cluster')[profile_cols].mean().round(2)
cluster_counts = df['cluster'].value_counts().sort_index()

print("\n=== Taille des segments ===")
print(cluster_counts.to_string())

print("\n=== Profil (moyennes par cluster) ===")
display(cluster_profile) if 'display' in globals() else print(cluster_profile)

# 9) Visualisation 2D (PCA) pour inspection visuelle
pca = PCA(n_components=2, random_state=RANDOM_STATE)
X_pca = pca.fit_transform(X_proc)

plt.figure()
for cl in sorted(np.unique(labels)):
    idx = labels == cl
    plt.scatter(X_pca[idx,0], X_pca[idx,1], s=20, label=f"Cluster {cl}", alpha=0.7)
plt.title("Clusters K-means (projection PCA 2D)")
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.legend()
plt.grid(True)
plt.show()

# 10) Sauvegarde optionnelle: DataFrame enrichi
out_name = "Camp_Market_with_clusters.csv"
try:
    df.to_csv(out_name, index=False, sep=';')
    print(f"[INFO] Fichier sauvegardé: {out_name}")
except Exception as e:
    print(f"[WARN] Sauvegarde non effectuée ({e}).")

# Conseils d’interprétation (rappel):
# - Recency plus faible = client actif récemment
# - TotalMnt/AvgSpendPerPurchase = valeur client
# - NumWebVisitsMonth élevé avec peu d’achats => trafic non converti
# - CampaignsAccepted/Response = sensibilité aux campagnes
# Utilise cluster_profile pour nommer tes segments (ex: “Grands acheteurs fidèles”, “Trafic web non converti”, etc.)
# === Fin du bloc K-means ===
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# -------------------------------------------------------------------
# 0) Chargement des données
# -------------------------------------------------------------------
if 'df' not in globals():
    csv_candidates = ['Camp_Market.csv', './Camp_Market.csv', '/mnt/data/Camp_Market.csv']
    csv_path = next((p for p in csv_candidates if os.path.exists(p)), None)
    if csv_path is None:
        raise FileNotFoundError("Place 'Camp_Market.csv' à la racine du notebook.")
    df = pd.read_csv(csv_path, sep=';')

print(f"[INFO] Données chargées: df.shape={df.shape}")

# -------------------------------------------------------------------
# 1) Feature engineering minimal (robuste et interprétable)
# -------------------------------------------------------------------
df = df.copy()

# Age (année de ref ≈ 2014 dans ce jeu)
if 'Age' not in df.columns:
    df['Age'] = 2014 - df['Year_Birth']

# Enfants à domicile
df['Children'] = df[['Kidhome','Teenhome']].sum(axis=1)

# Montants & comportements
amount_cols = [c for c in ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds'] if c in df.columns]
purchase_cols = [c for c in ['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases'] if c in df.columns]

df['TotalMnt'] = df[amount_cols].sum(axis=1)
df['TotalPurchases'] = df[purchase_cols].sum(axis=1)

# Visites web / Recency / Income: déjà présents dans le dataset classique
for col in ['Income','Recency','NumWebVisitsMonth']:
    if col not in df.columns:
        raise ValueError(f"Colonne attendue manquante: {col}")

# Engagement historique (nb campagnes déjà acceptées AVANT la variable cible)
camp_cols = ['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']
present_camps = [c for c in camp_cols if c in df.columns]
if len(present_camps) < 6:
    raise ValueError("Il manque des colonnes de campagne dans le DataFrame.")

# -------------------------------------------------------------------
# 2) Fonctions utilitaires d'analyse
# -------------------------------------------------------------------
def cohen_d(x_accept, x_refuse):
    # Taille d'effet, utile pour juger l'ampleur de la différence
    x1 = x_accept.dropna().values
    x0 = x_refuse.dropna().values
    if len(x1) < 3 or len(x0) < 3:
        return np.nan
    m1, m0 = x1.mean(), x0.mean()
    s1, s0 = x1.std(ddof=1), x0.std(ddof=1)
    # Pooled SD
    n1, n0 = len(x1), len(x0)
    sp = np.sqrt(((n1-1)*s1**2 + (n0-1)*s0**2) / (n1+n0-2))
    if sp == 0:
        return np.nan
    return (m1 - m0) / sp

def profile_table(df, target_col, feature_list):
    # Tableau (acceptés vs non) + diff + Cohen's d
    acc = df[df[target_col] == 1]
    ref = df[df[target_col] == 0]
    rows = []
    for f in feature_list:
        m1 = acc[f].mean()
        m0 = ref[f].mean()
        diff = m1 - m0
        d = cohen_d(acc[f], ref[f])
        rows.append([f, m1, m0, diff, d])
    out = pd.DataFrame(rows, columns=['Feature','Mean_Accept','Mean_Refuse','Diff','Cohen_d']).set_index('Feature')
    return out.round(3), len(acc), len(ref)

def simple_logit_importance(df, target_col, feature_list, random_state=42):
    # Logit rapide avec standardisation + imputation médiane
    X = df[feature_list].copy()
    y = df[target_col].astype(int).copy()
    pipe = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler()),
        ('logit', LogisticRegression(max_iter=200, random_state=random_state))
    ])
    pipe.fit(X, y)
    coefs = pipe.named_steps['logit'].coef_[0]
    imp = pd.Series(coefs, index=feature_list).sort_values(key=np.abs, ascending=False)
    return imp

def quintile_stratified_rates(df, target_col, strat_col, q=5):
    # Taux d'acceptation selon les quintiles d'une variable continue (ex: Recency, Income, TotalMnt)
    s = pd.qcut(df[strat_col].rank(method='first'), q=q, labels=[f"Q{i}" for i in range(1,q+1)])
    rates = df.groupby(s)[target_col].mean().rename('accept_rate').mul(100)
    sizes = df.groupby(s)[target_col].size().rename('n')
    return pd.concat([rates, sizes], axis=1)

def plot_quintiles_comp(qdf2, qdfR, label, figsize=(8,5)):
    plt.figure(figsize=figsize)
    plt.plot(qdf2.index.astype(str), qdf2['accept_rate'], marker='o', label='Cmp2')
    plt.plot(qdfR.index.astype(str), qdfR['accept_rate'], marker='o', label='Response')
    plt.title(f"Taux d'acceptation par quintiles de {label}")
    plt.ylabel("Taux d'acceptation (%)")
    plt.xlabel(f"Quintiles de {label} (Q1 bas ←→ Q5 haut)")
    plt.grid(True); plt.legend(); plt.tight_layout()
    plt.show()

# -------------------------------------------------------------------
# 3) Variables pour l'analyse
# -------------------------------------------------------------------
features = [
    'Age','Income','Children','Recency',
    'TotalMnt','TotalPurchases','NumWebVisitsMonth'
]

# Sécurité: convertir en numérique si besoin
for c in features:
    df[c] = pd.to_numeric(df[c], errors='coerce')

# Engagement passé AVANT chaque cible:
df['Accepted_before_Cmp2'] = df[['AcceptedCmp1']].sum(axis=1)
df['Accepted_before_Response'] = df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']].sum(axis=1)

# -------------------------------------------------------------------
# 4) Vue d'ensemble des taux (contrôle)
# -------------------------------------------------------------------
rates = (df[['AcceptedCmp2','Response']].mean()*100).round(2)
n_total = len(df)
print("=== Taux d'acceptation globaux ===")
print(f"AcceptedCmp2 : {rates['AcceptedCmp2']}%  (n={int(df['AcceptedCmp2'].sum())}/{n_total})")
print(f"Response     : {rates['Response']}%  (n={int(df['Response'].sum())}/{n_total})")

plt.figure(figsize=(6,4))
plt.bar(['AcceptedCmp2','Response'], [rates['AcceptedCmp2'], rates['Response']], edgecolor='black')
plt.title("Taux d'acceptation (global)")
plt.ylabel("Taux (%)"); plt.ylim(0, max(20, rates['Response']*1.3))
for i,v in enumerate([rates['AcceptedCmp2'], rates['Response']]):
    plt.text(i, v+0.4, f"{v:.2f}%", ha='center')
plt.tight_layout(); plt.show()

# -------------------------------------------------------------------
# 5) Profils : acceptés vs non (moyennes, différences, taille d'effet)
# -------------------------------------------------------------------
print("\n=== Profils (Accepted vs Refused) — Cmp2 ===")
prof2, n2_acc, n2_ref = profile_table(df, 'AcceptedCmp2', features)
display(prof2)

print("\n=== Profils (Accepted vs Refused) — Response ===")
profR, nR_acc, nR_ref = profile_table(df, 'Response', features)
display(profR)

# -------------------------------------------------------------------
# 6) Importances (logit) — Quels drivers d'acceptation ?
# -------------------------------------------------------------------
imp2 = simple_logit_importance(df, 'AcceptedCmp2', features)
impR = simple_logit_importance(df, 'Response', features)

print("\n=== Importance (logit standardisé) — Cmp2 ===")
display(imp2.to_frame('coef').round(3))
print("\n=== Importance (logit standardisé) — Response ===")
display(impR.to_frame('coef').round(3))

# -------------------------------------------------------------------
# 7) Stratifications explicatives (Recency / Income / Valeur client)
# -------------------------------------------------------------------
q2_rec = quintile_stratified_rates(df, 'AcceptedCmp2', 'Recency', q=5)
qR_rec = quintile_stratified_rates(df, 'Response', 'Recency', q=5)
print("\n=== Taux par quintiles de Recency (plus petit = plus récent) ===")
display(pd.concat([q2_rec.add_prefix('Cmp2_'), qR_rec.add_prefix('Resp_')], axis=1))
plot_quintiles_comp(q2_rec, qR_rec, label="Recency")

q2_inc = quintile_stratified_rates(df, 'AcceptedCmp2', 'Income', q=5)
qR_inc = quintile_stratified_rates(df, 'Response', 'Income', q=5)
print("\n=== Taux par quintiles d'Income ===")
display(pd.concat([q2_inc.add_prefix('Cmp2_'), qR_inc.add_prefix('Resp_')], axis=1))
plot_quintiles_comp(q2_inc, qR_inc, label="Income")

q2_val = quintile_stratified_rates(df, 'AcceptedCmp2', 'TotalMnt', q=5)
qR_val = quintile_stratified_rates(df, 'Response', 'TotalMnt', q=5)
print("\n=== Taux par quintiles de Valeur client (TotalMnt) ===")
display(pd.concat([q2_val.add_prefix('Cmp2_'), qR_val.add_prefix('Resp_')], axis=1))
plot_quintiles_comp(q2_val, qR_val, label="Valeur client (TotalMnt)")

# -------------------------------------------------------------------
# 8) Effet d'acculturation / qualification de base (répondants précédents)
# -------------------------------------------------------------------
# Taux d'acceptation selon le nb de campagnes déjà acceptées avant la cible
def rate_by_prior_accepts(df, prior_col, target_col, max_cap=5):
    g = df.groupby(np.minimum(df[prior_col], max_cap))[target_col].mean().mul(100)
    n = df.groupby(np.minimum(df[prior_col], max_cap))[target_col].size()
    out = pd.concat([g.rename('accept_rate_%'), n.rename('n')], axis=1)
    out.index.name = f"Nb acceptations préalables (cap à {max_cap}+)"
    return out

tab2_prior = rate_by_prior_accepts(df, 'Accepted_before_Cmp2', 'AcceptedCmp2', max_cap=1)  # avant Cmp2: seulement Cmp1
tabR_prior = rate_by_prior_accepts(df, 'Accepted_before_Response', 'Response', max_cap=5)  # avant Response: Cmp1..5

print("\n=== Taux selon acceptations préalables — Cmp2 ===")
display(tab2_prior)
print("\n=== Taux selon acceptations préalables — Response ===")
display(tabR_prior)

# Visualisation
plt.figure(figsize=(7,4))
plt.plot(tab2_prior.index.astype(str), tab2_prior['accept_rate_%'], marker='o', label='Cmp2')
plt.plot(tabR_prior.index.astype(str), tabR_prior['accept_rate_%'], marker='o', label='Response')
plt.title("Effet de l'historique de réponses (qualification de base)")
plt.ylabel("Taux d'acceptation (%)"); plt.xlabel("Nb de campagnes déjà acceptées")
plt.grid(True); plt.legend(); plt.tight_layout(); plt.show()

# -------------------------------------------------------------------
# 9) Résumé interprétatif (guidelines) en console
# -------------------------------------------------------------------
def sign(x): 
    return "↑" if x>0 else ("↓" if x<0 else "≈")

def summarize(prof_table, label):
    print(f"\n--- Lecture rapide : {label} ---")
    for feat in ['Recency','Income','TotalMnt','NumWebVisitsMonth','Age','Children','TotalPurchases']:
        if feat in prof_table.index:
            diff = prof_table.loc[feat,'Diff']
            dval = prof_table.loc[feat,'Cohen_d']
            print(f"{feat:>18s} : acceptants {sign(diff)}  vs non ({diff:+.2f}; d={dval:.2f})")

summarize(prof2, "Cmp2 (faible taux)")
summarize(profR, "Response (taux élevé)")

print("\n[NOTE] Interprétation type à adapter selon les sorties obtenues :")
print("- Si pour Cmp2 les acceptants ont Recency plus élevée (moins récents), Income/TotalMnt plus faibles,")
print("  et que l'effet de l'historique (Accepted_before_Cmp2) est faible → ciblage froid / offre mal adaptée.")
print("- Si pour Response les acceptants sont plus récents (Recency plus basse), plus hauts revenus / valeur,")
print("  et que le taux croît nettement avec le nombre d'acceptations préalables → base mieux qualifiée,")
print("  segmentation/pression marketing plus pertinentes à la fin du cycle de campagnes.")
# === Fin du bloc ===
```
